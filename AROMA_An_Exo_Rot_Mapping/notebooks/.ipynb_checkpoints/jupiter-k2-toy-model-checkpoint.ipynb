```python
import matplotlib.pyplot as plt
import numpy as np
from astropy import units as u
import cv2
from mayavi import mlab
from tvtk.api import tvtk # python wrappers for the C++ vtk ecosystem

def find_nearest(array, value):
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    return idx

import os
from os.path import join
array = os.path.abspath('').split('/')
homedir = '/'
for i in range(1,7):
   homedir = join(homedir, array[i])
plotPath = join(homedir, 'plots')

#=================================================
# Prepare the Cassini wind speed plot-based figure:

cassini_filename = '/Users/nguyendat/Documents/GitHub/aroma/AROMA_An_Exo_Rot_Mapping/data/Cassini_jupiter_BetterSampling.csv'
cassini = np.genfromtxt(cassini_filename,delimiter=",",dtype=np.double)
wind=np.array(cassini[:,0])
latitude=np.array(cassini[:,1])

wind_sorted=wind[np.argsort(latitude)]
latitude_sorted=latitude[np.argsort(latitude)]

wind=wind_sorted
latitude=latitude_sorted

#=================================================

imgsize=500
xcen=(500.0-1.)/2.0
ycen=(500.0-1.)/2.0
```

## 2D MAP

### NORMAL CASE: INCLUDE EQUATORIAL WINDS

```python
#=================================================
imgsize=500
xcen=(500.0-1.)/2.0
ycen=(500.0-1.)/2.0

latitudes=np.zeros((imgsize,imgsize))
windspeed=np.zeros((imgsize,imgsize))
periodimg=np.zeros((imgsize,imgsize))

latitudes[:]=np.NaN
windspeed[:]=np.NaN
periodimg[:]=np.NaN

radius=245. # Radius of jupiter in the synthetic image

Rjup=u.R_jup.to('m')

for xx in range(0,imgsize-1):
    for yy in range(0,imgsize-1):
        #Determine if outside the disk?
        # if (xx-xcen)**2+(yy-ycen)**2 >= (radius)**2 or (yy < 353 and yy > 146):
        if (xx-xcen)**2+(yy-ycen)**2 >= (radius)**2:
            latitudes[yy,xx]=np.nan
            windspeed[yy,xx]=np.nan
            periodimg[yy,xx]=np.nan
        else:
            lat=np.degrees(np.arcsin(-(yy-ycen)/radius))
            #latitude=(np.arcsin((yy-ycen)/radius))
            latitudes[yy,xx]=lat
            windspeed[yy,xx]=wind[find_nearest(latitude,lat)]
            rotspeed=2.0*np.pi*Rjup*np.cos(np.arcsin(-(yy-ycen)/radius))/(9.*3600.+56*60.)
            circumference=2.0*np.pi*Rjup*np.cos(np.arcsin(-(yy-ycen)/radius))
            periodimg[yy,xx]=circumference/(rotspeed+windspeed[yy,xx])/3600.
```

### Run 3D Mayavi and Save

```python
# run_mayavi = True
run_mayavi = False

if run_mayavi:
    # create a figure window (and scene)
    fig = mlab.figure(size=(500, 500))

    # load and map the texture
    img = tvtk.JPEGReader()
    img.file_name = plotPath + '/Equirectangular_CassiniJupiter_map.jpeg'
    texture = tvtk.Texture(input_connection=img.output_port, interpolate=1)
    # (interpolate for a less raster appearance when zoomed in)

    # use a TexturedSphereSource, a.k.a. getting our hands dirty
    R = 1
    Nrad = 180

    # create the sphere source with a given radius and angular resolution
    sphere = tvtk.TexturedSphereSource(radius=R, theta_resolution=Nrad,
                                       phi_resolution=Nrad)

    # assemble rest of the pipeline, assign texture    
    sphere_mapper = tvtk.PolyDataMapper(input_connection=sphere.output_port)
    sphere_actor = tvtk.Actor(mapper=sphere_mapper, texture=texture)
    fig.scene.add_actor(sphere_actor)

    sphere_actor.property.lighting = False
    mlab.show()
    mlab.pitch(20)
```

```python
periodimg_3d_fd_0 = cv2.imread(plotPath + '/mayavi_plots/mayavi_jupiter_face.png', cv2.IMREAD_ANYDEPTH).astype(np.float32)
%matplotlib inline

# plt.close(), 
# plt.figure(figsize=(4,4))
# n, bins, patches = plt.hist(periodimg_3d[~np.isnan(periodimg_3d)], 60, density=True, facecolor='tab:orange', alpha=0.75)

# convert background to NaN
for xx in range(periodimg_3d_fd_0.shape[0]):
    for yy in range(periodimg_3d_fd_0.shape[1]):
        if periodimg_3d_fd_0[xx][yy] <= 0:
            periodimg_3d_fd_0[xx][yy] = np.NaN
            
# figure out linear scaling
pmin, pmax = np.nanmin(periodimg_fd), np.nanmax(periodimg_fd)
pmin_old, pmax_old = np.nanmin(periodimg_3d_fd_0), np.nanmax(periodimg_3d_fd_0)
a = [[pmin_old, 1], [pmax_old,1]]
b = [pmin, pmax]

coeff = np.linalg.solve(a,b)
periodimg_3d_fd_0_new = coeff[0] * periodimg_3d_fd_0 + coeff[1]

plt.figure(figsize=(4,4), dpi=120)
im = plt.imshow(periodimg_3d_fd_0_new)
plt.colorbar(im, shrink=1.00, orientation='vertical')

plt.figure(figsize=(4,3), dpi=120)
n, bins, patches = plt.hist(periodimg_3d_fd_0_new[~np.isnan(periodimg_3d_fd_0_new)], 60, density=True, facecolor='tab:orange', alpha=0.75)
plt.xlim(9.80, 10.0)
```

### Projection map

```python
#=================================================
imgsize=1000
xcen=(float(imgsize)-1.)/2.0
ycen=(float(imgsize)-1.)/2.0

run_again = True
if run_again:
    latitudes=np.zeros((imgsize,imgsize))
    windspeed_mc=np.zeros((imgsize,imgsize))
    periodimg_mc=np.zeros((imgsize,imgsize))

    latitudes[:]=np.NaN
    windspeed_mc[:]=np.NaN
    periodimg_mc[:]=np.NaN

    radius=imgsize/2 # Radius of jupiter in the synthetic image

    Rjup=u.R_jup.to('m')

    for xx in range(0,imgsize-1):
        for yy in range(0,imgsize-1):
            if (yy > 0.6666*1000/2 and yy < 1.3333*1000/2):
                latitudes[yy,xx]=np.nan
                windspeed_mc[yy,xx]=np.nan
                periodimg_mc[yy,xx]=np.nan
            else:
                lat=np.degrees(np.arcsin(-(yy-ycen)/radius))
                latitudes[yy,xx]=lat
                windspeed_mc[yy,xx]=wind[find_nearest(latitude,lat)]
                rotspeed=2.0*np.pi*Rjup*np.cos(np.arcsin(-(yy-ycen)/radius))/(9.*3600.+56*60.)
                circumference=2.0*np.pi*Rjup*np.cos(np.arcsin(-(yy-ycen)/radius))
                periodimg_mc[yy,xx]=circumference/(rotspeed+windspeed_mc[yy,xx])/3600.
```

```python
periodimg_3d_mc_0 = cv2.imread(plotPath + '/mayavi_plots/[NoEQ]mayavi_jupiter_face.png', cv2.IMREAD_ANYDEPTH).astype(np.float32)
%matplotlib inline

# plt.close(), 
# plt.figure(figsize=(4,4))
# n, bins, patches = plt.hist(periodimg_3d[~np.isnan(periodimg_3d)], 60, density=True, facecolor='tab:orange', alpha=0.75)

# convert background to NaN
for xx in range(periodimg_3d_mc_0.shape[0]):
    for yy in range(periodimg_3d_mc_0.shape[1]):
        if periodimg_3d_mc_0[xx][yy] <= 0 or periodimg_3d_mc_0[xx][yy] >= 255:
            periodimg_3d_mc_0[xx][yy] = np.NaN
            
# figure out linear scaling
pmin, pmax = np.nanmin(periodimg_mc), np.nanmax(periodimg_mc)
pmin_old, pmax_old = np.nanmin(periodimg_3d_mc_0), np.nanmax(periodimg_3d_mc_0)
a = [[pmin_old, 1], [pmax_old,1]]
b = [pmin, pmax]

coeff = np.linalg.solve(a,b)
periodimg_3d_mc_0_new = coeff[0] * periodimg_3d_mc_0 + coeff[1]

plt.figure(figsize=(4,4), dpi=120)
im = plt.imshow(periodimg_3d_mc_0_new)
plt.colorbar(im, shrink=1.00, orientation='vertical')

plt.figure(figsize=(4,3), dpi=120)
n, bins, patches = plt.hist(periodimg_3d_mc_0_new[~np.isnan(periodimg_3d_mc_0_new)], 60, density=True, facecolor='tab:orange', alpha=0.75)
plt.xlim(9.80, 10.0)
```
